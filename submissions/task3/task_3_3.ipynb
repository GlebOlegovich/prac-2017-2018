### Состояние склада на каждый день

Представим данные о продажах в более удобной форме:

```python
def sell_upgrade(sell):
    
    sell_upg = pd.DataFrame(columns=['date', 'apple', 'pen'])  
    apl = 0
    pen = 0
    pos = 0
    length = len(sell) - 1
       
    for i in range(length):
        letter = sell['sku_num'][i][6]
        if letter == 'a':
            apl = apl + 1
        elif letter == 'p':
            pen = pen + 1
        date = sell['date'][i]
        if date != sell['date'][i+1]:
            sell_upg.at[pos, 'apple'] = apl
            sell_upg.at[pos, 'pen'] = pen
            sell_upg.at[pos, 'date'] = date
            pos = pos + 1
            apl = 0
            pen = 0
                
    pos = pos + 1
    letter = sell['sku_num'][length][6]
    if letter == 'a':
        sell_upg.at[pos, 'apple'] = apl + 1
        sell_upg.at[pos, 'pen'] = pen
    elif letter == 'p':
        sell_upg.at[pos, 'apple'] = apl
        sell_upg.at[pos, 'pen'] = pen + 1
    sell_upg.at[pos, 'date'] = date
        
    return sell_upg
```

Для магазина b1

```python
stor_daily_b1 = pd.DataFrame(columns=['date', 'apple', 'pen'])

length_sell = len(sell_upg_b1) - 1
length_spl = len(spl_b1) - 1

curr_apple = 0
curr_pen = 0

for i in range(length_sell):
    
    curr_date = sell_upg_b1['date'][i]
    for j in range(length_spl):
        if spl_b1['date'][j] == curr_date:
            curr_apple = curr_apple + spl_b1['apple'][j]
            curr_pen = curr_pen + spl_b1['pen'][j]
            
    curr_apple = curr_apple - sell_upg_b1['apple'][i]
    curr_pen  = curr_pen - sell_upg_b1['pen'][i]
   
    stor_daily_b1.at[i, 'apple'] = curr_apple
    stor_daily_b1.at[i, 'pen'] = curr_pen
    stor_daily_b1.at[i, 'date'] = curr_date
    
stor_daily_b1.head(n=15)
```
