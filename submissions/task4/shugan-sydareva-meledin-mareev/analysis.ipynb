```python
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sns
rcParams['figure.figsize'] = 12, 9
np.seterr(divide='ignore', invalid='ignore')
sns.set()
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2})
sns.set_palette(sns.xkcd_palette(['red', 'purple', 'dark blue', 'grey']))

data = pd.read_csv('production-data.csv', index_col = False)

data2 = data.groupby('supplier')['produced', 'defects'].sum()
print(data2)
data2.plot(kind = 'bar')
plt.title('Total produced and defects')
plt.legend(loc='best')
plt.ylabel('Number of weapons')
plt.show()

data2['defects/produced'] = data2.apply(lambda row: row.defects/row.produced, axis=1)
data2['defects/produced'].plot(kind = 'bar')
plt.title('Defects/Produced')
plt.legend(loc='best')
plt.show()

data2 = data.groupby(['supplier', 'report.date'])['defects'].mean()
plt.plot(data2['harpy.co'], label = 'harpy.co')
plt.plot(data2['westeros.inc'],label = 'westeros.inc')
plt.title('Average number of defected weapons reported during time')
plt.xlabel('Report date')
plt.ylabel('Average number of defects')
plt.legend(loc='best')
plt.show()

data2 = data.groupby(['supplier', 'report.date'])['defects'].sum()
plt.plot(data2['harpy.co'], label = 'harpy.co')
plt.plot(data2['westeros.inc'],label = 'westeros.inc')
plt.title('Total numbers of defected weapons reported during time')
plt.xlabel('Report date')
plt.ylabel('Total number of defects')
plt.legend(loc='best')
plt.show()

data2 = data.groupby(['supplier', 'production.date'])['defects'].mean()
data2['harpy.co'].plot(label = 'harpy.co')
data2['westeros.inc'].plot(label = 'westeros.inc')
plt.title('Average number of defects during production')
plt.ylabel('Average number of defects')
plt.legend(loc='best')
plt.show()

data2 = data.groupby(['supplier', 'production.date'])['defects'].sum()
data2['harpy.co'].plot(label = 'harpy.co')
data2['westeros.inc'].plot(label = 'westeros.inc')
plt.title('Total number of defects during production')
plt.ylabel('Total number of defects')
plt.legend(loc='best')
plt.show()

data2 = data.groupby(['supplier', 'unsullen.id'])['defects'].sum()
data2['harpy.co'].plot(label = 'harpy.co')
data2['westeros.inc'].plot(label = 'westeros.inc')
plt.title('Total number of defects (unsullen statistic)')
plt.ylabel('Total number of defects')
plt.legend(loc='best')
plt.show()


data2 = data.groupby(['supplier', 'production.date'])['produced'].sum()
data2['harpy.co'].plot(label = 'harpy.co')
data2['westeros.inc'].plot(label = 'westeros.inc')
plt.ylabel('Total number of weapons produced')
plt.title('Total production during time')
plt.legend(loc='best')
plt.show()


data2 = data.groupby(['supplier', 'unsullen.id'])['produced'].sum()
data2['harpy.co'].plot(label = 'harpy.co')
data2['westeros.inc'].plot(label = 'westeros.inc')
plt.title('Total production (unsullen statistic)')
plt.ylabel('Weapons produced')
plt.legend(loc='best')
plt.show()


data2 = data.groupby(['supplier', 'unsullen.id'])['produced', 'defects'].sum()
data2['produced - defects'] = data2.apply(lambda row: row.produced - row.defects, axis=1)
data2 = data2.groupby(['supplier', 'unsullen.id'])['produced - defects'].sum()
data2['harpy.co'].plot(label = 'harpy.co')
data2['westeros.inc'].plot(label = 'westeros.inc')
plt.title('Total number of not broken weapons (unsullen statistic)')
plt.ylabel('Weapons produced - Defects')
plt.legend(loc='best')
plt.show()

data2 = data.groupby(['supplier', 'production.date'])['produced', 'defects'].sum()
data2['produced - defects'] = data2.apply(lambda row: row.produced - row.defects, axis=1)
data2 = data2.groupby(['supplier', 'production.date'])['produced - defects'].sum()
data2['harpy.co'].plot(label = 'harpy.co')
data2['westeros.inc'].plot(label = 'westeros.inc')
plt.title('Total numbers of not broken weapons during production')
plt.ylabel('Weapons produced - Defects')
plt.legend(loc='best')
plt.show()

data2 = data.groupby('supplier')['produced', 'defects'].sum()
data2['produced-defects'] = data2.apply(lambda row: row.produced - row.defects, axis=1)
data2['produced-defects'].plot(kind = 'bar')
plt.title('Total number of not broken weapons')
plt.legend(loc='best')
plt.show()
```
